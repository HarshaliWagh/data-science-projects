```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
```

## Data Cleaning

```python
credit = pd.read_excel('default of credit card clients.xls')
```

```python
credit.head(10)
```

```python
credit.nunique()
```

```python
credit.columns
```

```python
credit.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
credit.head()
```

```python
credit.rename(columns=credit.iloc[0], inplace = True)
credit.drop(credit.index[0], inplace = True)
```

```python
credit.head()
```

```python
credit.columns
```

```python
credit['SEX'].replace([1, 2,], ['Male', 'Female'],inplace=True)
```

```python
credit.EDUCATION.value_counts()
```

```python
credit['EDUCATION'].replace([0,1,2,3,4,5,6], ['N/A', 'Graduate school', 'University', 'High school', 'Others', 'N/A', 'N/A'],inplace=True)
```

```python
credit.MARRIAGE.value_counts()
```

```python
credit['MARRIAGE'].replace([0,1,2,3], ['Others','Married','Single','Others'],inplace=True)
```

```python
credit.head(10)
```

```python
credit.shape
```

```python
credit.describe()
```

```python
credit.isnull().sum()
```

## Data Visualisation and Analysis

```python
credit['EDUCATION'].hist()
pt.xlabel('EDUCATION')
pt.ylabel('COUNT')
pt.title('EDUCATION v/s COUNT')
```
