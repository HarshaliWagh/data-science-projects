```python
import pandas as pd
import numpy as np
```

```python
obesity = pd.read_csv("obesity.csv")
```

```python
obesity.head()
```

***The data is highly uncleaned. Let's start it cleaning***

```python
obesity.nunique()
```

```python
obesity.shape
```

***The first two rows looks like repeating the same information. Confirm it with unique function.***

```python
obesity.iloc[0, :].unique()
```

```python
obesity.iloc[1, :].unique()
```

```python
obesity.iloc[2, :].unique()
```

***All columns in first two rows have the same values. The third row is the gender. 2016.1 is male, 2016.2 is female and 2016 is the average (both sex) and this is valid for all other years as well. So the first three rows are redundant and we can just drop them.***

```python
obesity.drop([0,1,2], axis=0, inplace=True)
```

```python
obesity.head()
```

```python
obesity.reset_index(drop=True, inplace=True)
```

```python
obesity.drop(obesity.columns[1::3], axis=1, inplace=True)
```

```python
obesity.head()
```

***Considering only the average value of the year and ignoring the range***

```python
for i in range(1,85):
    obesity.iloc[:,i] = obesity.iloc[:,i].str.split(expand=True)[0]
obesity.head()
```

***Converting it into narrow form by transposing columns and rows***

```python
obesity= obesity.melt(id_vars=['Unnamed: 0'], value_name='obesity_rate')
obesity.head()
```

***Creating gender column***

```python
obesity[['year','gender']] = obesity.iloc[:,1].str.split('.', expand=True)
obesity.head()
```

```python
obesity.drop('variable', axis=1, inplace=True)
gender = {'1':'male', '2':'female'}
obesity.gender.replace(gender, inplace=True)
obesity.rename(columns={'Unnamed: 0':'country'}, inplace=True)
obesity.head()
```

```python
obesity.tail()
```

***It looks like the data is cleaned properly***
